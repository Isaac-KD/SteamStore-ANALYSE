```python
def clean_per_column(df, column_name):
    if column_name in {"app_id", "description_courte", "date_de_sortie"}:
        return df # rien a faire
    if column_name in {"nom", "image", "description_detaillee"}:
        return df.drop([column_name], axis=1) # on drop la colonne name, image, description_detaillee
    if column_name == "type":
        return df[df[column_name] == "jeu"] # on drop les dlcs
    if column_name == "est_gratuit":
        return df[df[column_name] == False] # on drop les jeux gratuits
    if column_name == "developpeurs":
        return df.apply(lambda x : '_'.join(sorted(x)))
    else:
        return df
    

def clean_df(df):
    df_clean = df.copy()
    columns = df_clean.columns
    for c in columns:
        df_clean = clean_per_column(df_clean, c)
    return df_clean
    
```

```python
def clean_per_column(df, column_name):
    # Sécurité : si la colonne a déjà été supprimée par une étape précédente
    if column_name not in df.columns:
        return df

    # --- COLONNES À CONSERVER TELLES QUELLES ---
    if column_name in {"app_id", "description_courte"}:
        return df

    # --- COLONNES À SUPPRIMER (Inutiles pour l'analyse) ---
    # Ajout de 'configuration_requise' car trop complexe à parser proprement dans un premier temps
    if column_name in {"nom", "image", "description_detaillee", "configuration_requise", "langues"}:
        return df.drop(columns=[column_name])

    # --- FILTRAGE DES LIGNES ---
    if column_name == "type":
        # On ne garde que les jeux, puis on supprime la colonne devenue inutile (car n'ayant qu'une seule valeur)
        df = df[df[column_name] == "jeu"].copy()
        return df.drop(columns=[column_name])

    if column_name == "est_gratuit":
        # On ne garde que les jeux payants, puis on supprime la colonne
        df = df[df[column_name] == False].copy()
        return df.drop(columns=[column_name])

    # --- APLATISSEMENT DES LISTES (Arrays -> Strings) ---
    # Transforme ['Dev A', 'Dev B'] en "Dev A, Dev B"
    list_columns = ["developpeurs", "editeurs", "genres", "categories", "tags_utilisateurs", "plateformes"]
    if column_name in list_columns:
        # On vérifie si c'est bien une liste avant de join, sinon on renvoie une chaîne vide ou NaN
        df[column_name] = df[column_name].apply(lambda x: ', '.join(sorted(x)) if isinstance(x, list) else np.nan)
        return df

    # --- CONVERSIONS DE TYPES ---
    if column_name == "date_de_sortie":
        # Force la conversion en datetime, les erreurs deviennent NaT (Not a Time)
        df[column_name] = pd.to_datetime(df[column_name], errors='coerce')
        return df

    if column_name == "support_manette":
        # Remplit les valeurs nulles par 'none' pour avoir une catégorie propre
        df[column_name] = df[column_name].fillna('none')
        return df

    if column_name == "franchise":
         df[column_name] = df[column_name].fillna('Unknown')
         return df

    # --- EXTRACTION D'OBJETS IMBRIQUÉS ---
    if column_name == "evaluations":
        # Extraction des sous-champs importants en nouvelles colonnes
        # Utilisation de .get() pour éviter les erreurs si la clé manque
        df['metacritic'] = df[column_name].apply(lambda x: x.get('metacritic_score') if isinstance(x, dict) else np.nan)
        df['recommandations'] = df[column_name].apply(lambda x: x.get('recommandations_total') if isinstance(x, dict) else 0)
        # Pour les avis utilisateurs imbriqués
        df['avis_positifs_pct'] = df[column_name].apply(lambda x: x.get('avis_utilisateurs', {}).get('pourcentage_positif') if isinstance(x, dict) else np.nan)
        df['avis_total'] = df[column_name].apply(lambda x: x.get('avis_utilisateurs', {}).get('total') if isinstance(x, dict) else 0)
        df = df[df["avis_total"] >= 10]
        return df.drop(columns=[column_name])

    if column_name == "donnees_commerciales":
        df['prix_final'] = df[column_name].apply(lambda x: x.get('prix_final') if isinstance(x, dict) else np.nan)
        df = df[df["donnees_commerciales"].apply(lambda x: x["devise"]) == "EUR"]
        # df['reduction_pct'] = df[column_name].apply(lambda x: x.get('pourcentage_reduction') if isinstance(x, dict) else 0)
        # # On compte juste le nombre de DLCs au lieu de garder la liste d'IDs
        # df['nb_dlcs'] = df[column_name].apply(lambda x: len(x.get('dlcs', [])) if isinstance(x, dict) and 'dlcs' in x else 0)

        return df.drop(columns=[column_name])

    if column_name == "contenu_et_fonctionnalites":
        df['nb_succes'] = df[column_name].apply(lambda x: x.get('nombre_succes') if isinstance(x, dict) else 0)
        return df.drop(columns=[column_name])

    return df

def clean_df(df):
    df_clean = df.copy()
    # IMPORTANT : On itère sur une COPIE de la liste des colonnes initiales.
    # Car si on supprime des colonnes pendant la boucle, l'itérateur peut se perdre.
    initial_columns = list(df_clean.columns)
    for c in initial_columns:
        df_clean = clean_per_column(df_clean, c)

    # Nettoyage final : suppression des doublons éventuels basés sur l'ID
    if 'app_id' in df_clean.columns:
        df_clean = df_clean.drop_duplicates(subset=['app_id'])

    return df_clean.set_index('app_id')
```

```python

def visualize_data(df_input):
    """
    Génère des visualisations pour un DataFrame nettoyé, en excluant temporairement 
    les outliers extrêmes pour une meilleure lisibilité des graphiques numériques.
    """
    # Copie pour ne pas modifier le df original hors de la fonction
    df = df_input.copy()
    
    sns.set_theme(style="whitegrid")
    
    # 1. Identification des types de colonnes
    num_cols = df.select_dtypes(include=[np.number]).columns
    cat_cols = df.select_dtypes(include=['object', 'category']).columns

    # --- CONFIGURATION DES EXCLUSIONS CATÉGORIELLES ---
    exclusions = {
        'genres': {'nudity', 'sexual content', 'violent', 'gore'},
        'developpeurs': {'llc', 'inc', 'inc.', 'ltd', 'ltd.', 's.r.l.', 'limited', 'corporation', 'corp.', 's.a.', 'co.', 'company'},
        'editeurs': {'llc', 'inc', 'inc.', 'ltd', 'ltd.', 's.r.l.', 'limited', 'corporation', 'corp.', 's.a.', 'co.', 'company'}
    }
    force_plot_cols = ['genres', 'tags_utilisateurs', 'categories', 'developpeurs', 'editeurs', 'plateformes']

    print("--- Résumé ---")
    print(df.describe())
    print("\n")

    # 2. Visualisation des variables NUMÉRIQUES (Zoom sur la partie centrale)
    for col in num_cols:
        # --- FILTRAGE DES OUTLIERS EXTRÊMES (Méthode IQR) ---
        # On calcule Q1, Q3 et l'IQR pour définir les bornes "normales"
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Définition des bornes pour filtrer les valeurs extrêmes (seuil classique de 1.5 * IQR)
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Création d'une série temporaire sans les valeurs extrêmes pour le plot
        # On filtre pour ne garder que les valeurs dans l'intervalle [lower_bound, upper_bound]
        col_filtered = df[col][(df[col] >= lower_bound) & (df[col] <= upper_bound)]
        
        n_dropped = len(df[col]) - len(col_filtered) # Nombre de valeurs exclues

        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        fig.suptitle(f'Analyse de : {col} (Zoom central - {n_dropped} outliers masqués)', fontsize=16)

        # Histogramme sur données filtrées
        sns.histplot(col_filtered, kde=True, ax=axes[0], color='skyblue')
        axes[0].set_title('Distribution (sans valeurs extrêmes)')
        axes[0].axvline(col_filtered.mean(), color='r', linestyle='--', label=f'Moyenne (filtrée): {col_filtered.mean():.2f}')
        axes[0].axvline(col_filtered.median(), color='g', linestyle='-', label=f'Médiane (filtrée): {col_filtered.median():.2f}')
        axes[0].legend()

        # Boxplot sur données filtrées
        sns.boxplot(x=col_filtered, ax=axes[1], color='lightgreen')
        axes[1].set_title('Boîte à moustaches (Zone centrale uniquement)')

        plt.tight_layout()
        plt.show()

    # 3. Visualisation des variables CATÉGORIELLES (inchangé)
    for col in cat_cols:
        if df[col].nunique() > 500 and col not in force_plot_cols: 
             continue

        plt.figure(figsize=(12, 8))
        current_exclusions = exclusions.get(col, set())

        if df[col].astype(str).str.contains(',').any():
            exploded = df[col].str.split(', ').explode()
            mask_valide = (
                exploded.str.strip().astype(bool) & 
                ~exploded.str.lower().str.strip().isin(current_exclusions)
            )
            exploded_clean = exploded[mask_valide]
            top_values = exploded_clean.value_counts().head(20)
            title_suffix = "(Valeurs séparées & filtrées)"
        else:
            mask_valide = ~df[col].str.lower().str.strip().isin(current_exclusions)
            top_values = df[col][mask_valide].value_counts().head(20)
            title_suffix = "(Filtrées)"

        sns.barplot(x=top_values.values, y=top_values.index, palette='viridis')
        plt.title(f'Top 20 : {col} {title_suffix}', fontsize=14)
        plt.xlabel("Nombre d'occurrences")
        plt.ylabel(col.capitalize())
        plt.tight_layout()
        plt.show()
```
